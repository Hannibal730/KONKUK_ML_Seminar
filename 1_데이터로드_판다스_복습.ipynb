# 3. 판다스 함수 복습
- iloc: 정수 기반 인덱싱
- loc: 라벨이름 기반 인덱싱

```python
#시작하는 줄부터 5줄 인쇄하기
print(train.head())
#옵션을 주지 않으면 기본 5줄 인쇄,
#옵션 주려면 (n) 괄호 안 n에 숫자 입력

print(train.head(3))

print(train.tail(4))
```

```python
# 열 추출하기
print(train['country'])

# 여러 열 추출하기
train[  ['continent', 'lifeExp']   ]


# 특정 열을 추출하여 변수에 담아둘 수도 있다. 특정 열만 변경하거나 인코딩할 때에 필요하다.
country=train['country']
print(country)
```

```python
# 한 행 추출
train.loc[0]
```

```python
# 여러 행 추출
train.loc[ [10,20,30] ]
```

```python
# 여러 행과 열 추출
train.loc[1:5, ['year', 'pop']]
```

```python
# 숫자로 데잍터를 추출하는  iloc
# iloc을 쓸 때는 행과 열을 모두 숫자로 슬라이싱/인덱싱해야 한다.
train.iloc[0:3, [0,5]]
```

# group by

```python
# 데이터프레임.groupby(기준이 되는 열)[계산할 대상인 열].함수

train.groupby('year')['lifeExp'].mean()
```

```python
# nunique(): 기준이 되는 열에 유니크한 값의 개수가 몇 개인지
train.groupby('continent')['lifeExp'].nunique()
```

```python
# value_counts: 각 항목의 개수
train.groupby('continent')['country'].value_counts()
```
