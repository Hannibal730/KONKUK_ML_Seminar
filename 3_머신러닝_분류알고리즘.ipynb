# 분류 알고리즘
- k-nn 알고리즘: 특정 데이터 인근의 데이터 k개를 보고서, 데이터의 종류를 결정.
- 의사결정나무

```python
import pandas as pd
#사이킷런이 가지고 있는 데이터를 가져와 사용(iris)
from sklearn.datasets import load_iris
```

```python
#붓꽃 데이터셋을 data라는 변수에 입력
data = load_iris()
data # 데이터는 데이터, 타켓은 라벨.
```

```python
x= pd.DataFrame(data.data, columns=data.feature_names) # 데이터에 피쳐네임이 자체 저장되어 있었는데, 그걸 열로 지정해줌.
y= pd.DataFrame(data.target, columns=['result']) # 라벨의 컬럼은 result로 지정해줌.
```

```python
x.head(5)
```

```python
y.tail(3)
```

```python
df=pd.concat([x,y],axis=1) # concat함수로 데이터프레임 붙이기.
df.head()
```

# 머신러닝 알고리즘 적용

```python
from sklearn.neighbors import KNeighborsClassifier

# knn 모델은 n_neighbors매개변수에 k값을 지정해줘야 함.
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x,y) # 학습시키는 fit함수, 데이터x, 라벨y
```

```python
# 학습시킨 x데이터로 knn에게 예측 시켜보기.
knn_predict=knn.predict(x) # 예측하는 predict함수
```

```python
# 머신러닝의 다양한 평가지표 중에 정확도를 사용.
from sklearn.metrics import accuracy_score

accuracy_score(y,knn_predict) # 정답 레이블인 y와 knn_predict의 결과를 비교하여 정확도 산출
```

# 의사결정나무로 동일한 데이터셋 학습

```python
from sklearn import tree
clf =tree.DecisionTreeClassifier()
clf.fit(x,y) # 데이터x와 라벨y를 fir함수로 학습시킴.
```

```python
clf_predict=clf.predict(x) # 그냥 clf.predict(x)만 써도 됨.
```

```python
accuracy_score(y,clf.predict(x)) # 라벨이랑 예측값을 비교
```

# 로지스틱 회귀 알고리즘을 가지고, 위와 같이 학습과 평가를 수행

```python
from sklearn.linear_model import LogisticRegression
```

- 이런식으로 라이브러리만 주어진다면, 위에서 했던 다른 알고리즘들처럼 비슷하게 적용가능.

- 복습.
1. 사이킷런에서 알고리즘과 어큐러시 스코어 임포트
2. 라이브러리를 인스턴스로 받아오고
3. 핏 함수, 데이터, 라벨 활용해서 학습시키기
4. 프레딕트 함수 써서 예측해보기
5. 어큐러시 스코어 함수, 라벨, 예측값 활용해서 정답률 계산

```python
log=LogisticRegression()
log.fit(x,y) # 훈련데이터 x
log_predict=log.predict(x) # 예측데이터 x..?  어라 그러네. fit 훈련 때 x 이미 썼는데, 예측할 때도 x 써버림;
accuracy=accuracy_score(y,log_predict)
print(f'정확도는 {accuracy}입니다.\n')
```

# 훈련데이터와 검증데이터 분리
- 사이킷런의 트레인 테스트 스플릿 함수를 활용하기

```python
len(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
# 랜덤스테이트: 트레인테스트스플릿으로 데이터를 랜덤분리할 때, 그 랜덤분리 결과의 고유번호
# 랜덤으로 하면 좋을텐데 굳이 고정시켜두는 이유: 논문검증, 보고서 작성...등등. 물론 생략해도 됨.

log2=LogisticRegression()
log2.fit(x_train,y_train)

log2_predict=log2.predict(x_test)
accuracy2=accuracy_score(y_test,log2_predict)
print(f'정확도는 {accuracy2}입니다.\n')
```
